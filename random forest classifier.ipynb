```python
data = pd.read_csv('final_merged_eth_gold_global_data_v1.csv', parse_dates=['Date'], index_col='Date')
data
```

```python
# closing price for next day

data["tomorrow"] = data['Close'].shift(-1)
data
```

```python
# The purpose of the function predict_fixed is to simplify and construct the training and prediction process of 
# machine learning models. The steps here are: Model training: use the provided training data train x train 
# and the target value train the y train to train the specified model. 
# Make predictions: use a model already trained on the test data test x test forecast on. 
# Combine result predictions and true target values: Combine the predicted results and the true target values 
# in the test set in one data bin for subsequent comparison and analysis
def predict_fixed(X_train, y_train, X_test, y_test, model):
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    preds = model.predict(X_test)

    # Combine the true target values and predictions into a dataframe
    preds_series = pd.Series(preds, index=X_test.index, name="prediction")
    combined = pd.concat([y_test, preds_series], axis=1)
    
    return combined

# Just as a demonstration, let's use this function again with the provided datasets
combined_results_demo = predict_fixed(X_train_extended, y_train, X_test_extended, y_test, model_extended)
combined_results_demo.head()
```

```python
# Define possible step values to test
steps_to_test = [30, 60, 90, 120, 150, 180, 210, 240, 270, 300]

# Store results for each step value
step_results = {}

# Run backtest for each step value and calculate its accuracy
for step in steps_to_test:
    backtest_result = backtest_improved(data, model_extended, extended_predictors, step=step)
    correct_predictions = (backtest_result["target"] == backtest_result["prediction"]).sum()
    accuracy = correct_predictions / len(backtest_result)
    step_results[step] = accuracy

step_results
```

```python
def compute_rolling(data):
    horizons = [2,7,60,365]
    new_predictors = ['Close', '^GSPC', '^NYA', '^DJI', "Open", "High", "Low", "Volume"]
    
    for horizon in horizons:
        rolling_averages = data.rolling(horizon, min_periods=1).mean()
        
        ratio_column = f"close_ratio_{horizon}"
        data[ratio_column] = data["Close"] / rolling_averages["Close"]
        
        rolling = data.rolling(horizon, closed='left', min_periods=1).mean()
        trend_column = f"trend_{horizon}"
        data[trend_column] = rolling["target"]
    
        new_predictors += [ratio_column, trend_column]
    return data, new_predictors

data, new_predictors = compute_rolling(data.copy())
```

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import precision_score

# Load the dataset
data = pd.read_csv('final_merged_eth_gold_global_data_v1.csv', parse_dates=['Date'], index_col='Date')

# Compute target variable
data["tomorrow"] = data['Close'].shift(-1)
data["target"] = (data["tomorrow"] > data["Close"]).astype(int)  # 1 if price goes up, 0 if goes down
data = data.dropna()

def compute_rolling(data):
    horizons = [2,7,60,365]
    new_predictors = ['Close', '^GSPC', '^NYA', '^DJI', "Open", "High", "Low", "Volume"]
    
    for horizon in horizons:
        rolling_averages = data.rolling(horizon, min_periods=1).mean()
        
        ratio_column = f"close_ratio_{horizon}"
        data[ratio_column] = data["Close"] / rolling_averages["Close"]
        
        rolling = data.rolling(horizon, closed='left', min_periods=1).mean()
        trend_column = f"trend_{horizon}"
        data[trend_column] = rolling["target"]
    
        new_predictors += [ratio_column, trend_column]
    return data, new_predictors

data, new_predictors = compute_rolling(data.copy())

# Splitting data for backtesting
train = data.iloc[:-200]
test = data.iloc[-200:]

def predict_fixed_best(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    preds_series = pd.Series(preds, index=X_test.index, name="prediction")
    combined = pd.concat([y_test, preds_series], axis=1)
    return combined

def backtest_improved(data, model, predictors, start=540, step=60):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train_data = data.iloc[0:i].copy()
        test_data = data.iloc[i:min(i+step, data.shape[0])].copy()
        
        X_train = train_data[predictors].fillna(train_data[predictors].mean())
        y_train = train_data["target"]
        X_test = test_data[predictors].fillna(test_data[predictors].mean())
        y_test = test_data["target"]
        
        predictions = predict_fixed_best(X_train, y_train, X_test, y_test, model)
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)

# Using the hyperparameters from GridSearchCV
optimized_model = XGBClassifier(
    learning_rate=0.1,
    n_estimators=200,
    max_depth=5,
    min_child_weight=5,
    gamma=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary:logistic",
    scale_pos_weight=1,
    seed=1
)

predictions = backtest_improved(data, optimized_model, new_predictors)
precision = precision_score(predictions["target"], predictions["prediction"])  
print(precision)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score

# Reducing the step size to speed up the backtesting process
predictions = backtest_improved(data, optimized_model, new_predictors, step=30)

# Calculating the performance metrics again
accuracy = (predictions["target"] == predictions["prediction"]).mean()
precision = precision_score(predictions["target"], predictions["prediction"])
recall = recall_score(predictions["target"], predictions["prediction"])
f1 = f1_score(predictions["target"], predictions["prediction"])

accuracy, precision, recall, f1
```
